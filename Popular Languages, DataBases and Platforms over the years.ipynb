## As we see, there are multiple values per person's entry seperated by ";". So to get better uderstanding,  we need to split these values and count them seperately. 

### For 2017

```python

df_2017_languages = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2017.HaveWorkedLanguage.dropna().tolist()).split(';')])
df_2017_languages.columns = ['languages']
df_2017_languages = df_2017_languages.value_counts().rename_axis('languages').reset_index(name='counts')

df_2017_languages
```

```python
df_2017_databases = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2017.HaveWorkedDatabase.dropna().tolist()).split(';')])
df_2017_databases.columns = ['databases']
df_2017_databases = df_2017_databases.value_counts().rename_axis('databases').reset_index(name='counts')
df_2017_databases
```

```python

df_2017_platforms = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2017.HaveWorkedPlatform.dropna().tolist()).split(';')])
df_2017_platforms.columns = ['platforms']
df_2017_platforms = df_2017_platforms.value_counts().rename_axis('platforms').reset_index(name='counts')
df_2017_platforms

```

### For 2018

```python


df_2018_languages = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2018.LanguageWorkedWith.dropna().tolist()).split(';')])
df_2018_languages.columns = ['languages']
df_2018_languages = df_2018_languages.value_counts().rename_axis('languages').reset_index(name='counts')

df_2018_databases = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2018.DatabaseWorkedWith.dropna().tolist()).split(';')])
df_2018_databases.columns = ['databases']
df_2018_databases = df_2018_databases.value_counts().rename_axis('databases').reset_index(name='counts')

df_2018_platforms = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2018.PlatformWorkedWith.dropna().tolist()).split(';')])
df_2018_platforms.columns = ['platforms']
df_2018_platforms = df_2018_platforms.value_counts().rename_axis('platforms').reset_index(name='counts')


```

### For 2019

```python


df_2019_languages = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2019.LanguageWorkedWith.dropna().tolist()).split(';')])
df_2019_languages.columns = ['languages']
df_2019_languages = df_2019_languages.value_counts().rename_axis('languages').reset_index(name='counts')

df_2019_databases = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2019.DatabaseWorkedWith.dropna().tolist()).split(';')])
df_2019_databases.columns = ['databases']
df_2019_databases = df_2019_databases.value_counts().rename_axis('databases').reset_index(name='counts')

df_2019_platforms = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2019.PlatformWorkedWith.dropna().tolist()).split(';')])
df_2019_platforms.columns = ['platforms']
df_2019_platforms = df_2019_platforms.value_counts().rename_axis('platforms').reset_index(name='counts')


```

### For 2020

```python


df_2020_languages = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2020.LanguageWorkedWith.dropna().tolist()).split(';')])
df_2020_languages.columns = ['languages']
df_2020_languages = df_2020_languages.value_counts().rename_axis('languages').reset_index(name='counts')

df_2020_databases = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2020.DatabaseWorkedWith.dropna().tolist()).split(';')])
df_2020_databases.columns = ['databases']
df_2020_databases = df_2020_databases.value_counts().rename_axis('databases').reset_index(name='counts')

df_2020_platforms = pd.DataFrame([x.strip(' ') for x in ';'.join(data_2020.PlatformWorkedWith.dropna().tolist()).split(';')])
df_2020_platforms.columns = ['platforms']
df_2020_platforms = df_2020_platforms.value_counts().rename_axis('platforms').reset_index(name='counts')


```

## Now lets combine al the dataframes over the years to get a better understanding.

```python
temp = pd.merge(df_2017_languages, df_2018_languages, on='languages')
temp = pd.merge(temp, df_2019_languages, on='languages')
language_trends = pd.merge(temp, df_2020_languages, on='languages')
language_trends.columns = ['languages','2017', '2018', '2019','2020']

language_trends

```

```python
temp = pd.merge(df_2017_databases, df_2018_databases, on='databases')
temp = pd.merge(temp, df_2019_databases, on='databases')
database_trends = pd.merge(temp, df_2020_databases, on='databases')
database_trends.columns = ['databases','2017', '2018', '2019','2020']

database_trends

```

```python

temp = pd.merge(df_2017_platforms, df_2018_platforms, on='platforms')
temp = pd.merge(temp, df_2019_platforms, on='platforms')
platform_trends = pd.merge(temp, df_2020_platforms, on='platforms')
platform_trends.columns = ['platforms','2017', '2018', '2019','2020']

platform_trends 

```

## Now lets do the fun part! PLOTTING

### Language Trends over the Year

```python
language_trends_norm = language_trends.copy()
language_trends_norm.iloc[:,1:] = (language_trends.iloc[:,1:]/language_trends.iloc[:,1:].sum())*100
language_trends_norm.index = language_trends_norm.languages
language_trends_norm.plot(title='Programming Languages in 2017, 2018 , 2019 and 2020',kind='bar', figsize=(13,7))
```

## DataBase Trends over the Year

```python
database_trends_norm = database_trends.copy()
database_trends_norm.iloc[:,1:] = (database_trends_norm.iloc[:,1:]/database_trends_norm.iloc[:,1:].sum())*100
database_trends_norm.index = database_trends_norm.databases
database_trends_norm.plot(title='Databases in 2017, 2018 , 2019 and 2020',kind='bar', figsize=(13,7))
```

## Platform Trends over the Year

```python
platform_trends_norm = platform_trends.copy()
platform_trends_norm.iloc[:,1:] = (platform_trends_norm.iloc[:,1:]/platform_trends_norm.iloc[:,1:].sum())*100
platform_trends_norm.index = platform_trends_norm.platforms
platform_trends_norm.plot(title='Platforms in 2017, 2018 , 2019 and 2020',kind='bar', figsize=(13,7))
```

## Results

In Languages, we see that the number or developers using Python, TypeScript, Go and Rust are on the rise while Javascript is still in the lead while Java and PHP are on a downward trend.

In Database trends, we see MySQL is still on the top but decreasing is market cap whereas PostgreSQL and MongoDb are rising.

In terms of Platforms, we see number of Android devs are rising while iOS are decreasing.
