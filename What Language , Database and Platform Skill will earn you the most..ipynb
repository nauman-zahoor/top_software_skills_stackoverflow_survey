## Lets first thershold the data based on quantiles such that the resulting skewness and distribution of all years is close to each other
## Purpose of doing this is to not add bias from small number of people with very large saleries as some of them would be present in one year and won't be present in other years. So lets keep only that data that has same distribution over the years

```python
# 2017

data_2017_s2 = data_2017_s[data_2017_s.Salary> data_2017_s.Salary.quantile(0.1) ]
data_2017_s2 = data_2017_s2[data_2017_s2.Salary< data_2017_s.Salary.quantile(1) ]

data_2017_s2 = data_2017_s2.reset_index(drop=True)

print('Skew = ',data_2017_s2.Salary.skew())
data_2017_s2.Salary.hist()
```

```python
#2018

data_2018_s2 = data_2018_s[data_2018_s.Salary> data_2018_s.Salary.quantile(0.1) ]
data_2018_s2 = data_2018_s2[data_2018_s2.Salary< data_2018_s.Salary.quantile(0.95) ]
data_2018_s2 = data_2018_s2.reset_index(drop=True)
print('Skew = ',data_2018_s2.Salary.skew())
data_2018_s2.Salary.hist()
```

```python
##### 2019

data_2019_s2 = data_2019_s[data_2019_s.Salary> data_2019_s.Salary.quantile(0.1) ]
data_2019_s2 = data_2019_s2[data_2019_s2.Salary< data_2019_s.Salary.quantile(0.91) ]
data_2019_s2 = data_2019_s2.reset_index(drop=True)
print('Skew = ',data_2019_s2.Salary.skew())
data_2019_s2.Salary.hist()
```

```python
#2020

data_2020_s2 = data_2020_s[data_2020_s.Salary> data_2020_s.Salary.quantile(0.1) ]
data_2020_s2 = data_2020_s2[data_2020_s2.Salary< data_2020_s2.Salary.quantile(0.93) ]
data_2020_s2 = data_2020_s2.reset_index(drop=True)
print('Skew = ',data_2020_s2.Salary.skew())
data_2020_s2.Salary.hist()
```

## Much Better. All of our data across the years has almost same distribution and skewness.

## Lets Start with languages and try to see what is trend with languages over time 

### We see that a person can select multiple languages so for our analysis, lets split all languages and assign each language the same salary the person has entered.

## 2017

```python

data_2017_s2
#Lets drop all rows where no languages are entered
data_2017_s3 = data_2017_s2.dropna(subset=['Languages']).reset_index(drop=True)
data_2017_s3.head()
```

```python
data = data_2017_s3
col = 'Languages'
sal = []
val = []

for i in range(len(data)):
    vals = data.loc[i,col].split(';')
    salary = data.loc[i,'Salary']
    
    for j in vals:
        val.append(j.strip())
        sal.append(salary)
df_new = pd.DataFrame(val)
df_new.columns = [col]
df_new.loc[:,'Salary'] = sal     

```

```python
data_2017_languages = df_new.groupby('Languages').mean().sort_values(by='Salary',ascending = False)
data_2017_languages.plot.bar(title='Salary per programming Language in 2017', figsize=(10,7))
```

## Now lets combine all the dataframes over the years to get a better understanding.

```python


# lets merge all years.
temp = pd.merge(data_2017_languages, data_2018_languages, on='Languages',how = 'right') 
temp = pd.merge(temp, data_2019_languages, on='Languages',how = 'right')
language_trends = pd.merge(temp, data_2020_languages, on='Languages',how = 'outer')
language_trends.columns = ['2017', '2018', '2019','2020']

language_trends
```


## let's Plot

```python

language_trends.plot(title='Salary vs Programming Languages in 2017, 2018 , 2019 and 2020',kind='bar', figsize=(13,8))
```

# Conclusion
## From the graph, we see that  PHP, R and VBA are on a downward trend indicating that those might be dead technologies. Clojure has been continously the highest paid programming language ovrer the years. Python , C, C++ and Java are on a continous rising trend.
